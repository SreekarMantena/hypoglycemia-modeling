```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lr = pd.read_csv('LR_thresh_mult_scor.csv')
xgb = pd.read_csv('XGB_thresh_mult_scor.csv')
```

```python
plt.figure(figsize = (6,6))
plt.plot(1 - lr['Specificity'], lr['Sensitivity'], label = 'ROC Curve of Logistic Regression (AUC = .78)', color = 'blue')
plt.plot(1 - xgb['Specificity'], xgb['Sensitivity'], label = 'ROC Curve of XGB (AUC = .8)', color = 'orange')
plt.plot([0, 1], [0, 1], color='black', lw=1, linestyle='--')
plt.title('ROC Curves for Prediction of Hypoglycemia on Test Dataset')
plt.xlabel('1 - Specificity')
plt.ylabel('Sensitivity')
plt.legend()
plt.savefig('roc_curves.jpg', dpi = 200)
```
